## **Vacances et bagage**
# pour resudre ce probleme on utilise la programmation dynamique

```python
def take_objects(obj_importance, obj_sizes, car_space):
    n = len(obj_importance)
    # on cree un tableau de taille (n+1)*(car_space+1) pour stocker les valeurs intermediaires
    # et on initialise les valeurs de la premiere ligne et de la premiere colonne a 0
    # car si on a pas d'objet ou pas d'espace on ne peut rien prendre
    table = [[0 for x in range(car_space + 1)] for x in range(n + 1)]

    # on parcourt les objets
    for i in range(n + 1):
        # on parcourt les espaces
        for j in range(car_space + 1):
            # si on a pas d'objet ou pas d'espace on ne peut rien prendre
            if i == 0 or j == 0:
                table[i][j] = 0
            # si l'objet actuel est plus petit que l'espace disponible
            # on prend le max entre le fait de prendre l'objet actuel ou de ne pas le prendre
            elif obj_sizes[i - 1] <= j:
                table[i][j] = max(obj_importance[i - 1] + table[i - 1][j - obj_sizes[i - 1]], table[i - 1][j])
            # sinon on ne prend pas l'objet actuel
            else:
                table[i][j] = table[i - 1][j]

    # on initialise les variables de retour
    opt = table[n][car_space]
    choices = []
    res = opt
    # on parcourt les objets
    for i in range(n, 0, -1):
        # si la valeur de l'objet actuel est differente de la valeur de l'objet precedent
        # on ajoute l'objet actuel a la liste des objets choisis
        if res != table[i - 1][car_space]:
            choices.append(i)
            res = res - obj_importance[i - 1]
            car_space = car_space - obj_sizes[i - 1]

    # on retourne les variables de retour
    return opt, choices
```

 Dans le contexte de la fonction take_objects, **opt** et **choices** ont les significations suivantes :


*    **opt :** Est la somme maximale des importances des objets que vous pouvez emporter sans dépasser l’espace total disponible dans la voiture.

*    **choices :** C’est une liste des indices des objets que vous avez choisis pour atteindre la valeur optimale. Par exemple, si choices est [2, 1], cela signifie que vous avez choisi les objets aux indices 2 et 1 de vos listes obj_importance et obj_sizes pour atteindre la valeur optimale. ***( l'indice commance par 1 et non par 0).***

 Donc, en résumé, opt est la meilleure importance totale que vous pouvez obtenir, et choices sont les objets que vous devez choisir pour atteindre cette importance.


```python
obj_importance = [3, 2, 4, 4]
obj_sizes = [2, 1, 3, 2]
car_space = 5
opt, choices = take_objects(obj_importance, obj_sizes, car_space)
print("Optimal value:", opt)
print("Chosen objects:", choices)
```

## **Sudoku**
### L’algorithme utilisé pour résoudre le Sudoku est le Backtracking

Le code est organisé comme suivant:


*   `est_valide(puzzle, ligne, colonne, num):` Cette fonction vérifie si un nombre donné `(num)` peut être placé à une position spécifique `(ligne, colonne)` dans le Sudoku` (puzzle)`.

*   `resoudre_sudoku(puzzle):` Cette fonction utilise un algorithme de backtracking pour résoudre le puzzle.

*   `afficher_sudoku(puzzle, puzzle_original):` Cette fonction est concernée par la visualisation de la matrice, nous utilisons Matplotlib, Les nombres qui ont été remplacés par l’algorithme sont affichés en rouge.

*   `sudoku(puzzle):` Cette fonction appel les fonctions précédentes.



```python
import matplotlib.pyplot as plt

def est_valide(puzzle, ligne, colonne, num):
    for x in range(9):
        if puzzle[ligne][x] == num:
            return False

    for x in range(9):
        if puzzle[x][colonne] == num:
            return False

    debut_ligne = ligne - ligne % 3
    debut_colonne = colonne - colonne % 3
    for i in range(3):
        for j in range(3):
            if puzzle[i + debut_ligne][j + debut_colonne] == num:
                return False
    return True

def resoudre_sudoku(puzzle):
    for ligne in range(9):
        for colonne in range(9):
            if puzzle[ligne][colonne] == '$':
                for num in range(1, 10):
                    if est_valide(puzzle, ligne, colonne, num):
                        puzzle[ligne][colonne] = num
                        if resoudre_sudoku(puzzle):
                            return True
                        puzzle[ligne][colonne] = '$'
                return False
    return True

def afficher_sudoku(puzzle, puzzle_original):
    fig, ax = plt.subplots()
# Set the background color
    ax.set_facecolor('lightgrey')
    # Draw the grid
    for i in range(10):
        if i % 3 == 0:
            ax.plot([i, i], [0,9], color='black', linewidth=2)
            ax.plot([0,9], [i,i], color='black', linewidth=2)
        else:
            ax.plot([i, i], [0,9], color='black', linewidth=1)
            ax.plot([0,9], [i,i], color='black', linewidth=1)

    # Add numbers and color the background of the cases
    for i in range(9):
        for j in range(9):
                color, facecolor = ('blue', 'lightblue') if puzzle[i][j] == puzzle_original[i][j] else ('red', 'pink')
                ax.text(j+0.5, 8.5-i, str(puzzle[i][j]), ha='center', va='center', color=color, bbox=dict(facecolor=facecolor))


    # Add title
    plt.title("Sudoku Puzzle", fontsize=20, pad=20)

    # Remove axis
    ax.axis('off')

    # Show the plot
    plt.show()

def sudoku(puzzle):
    puzzle_original = [row.copy() for row in puzzle]
    if(resoudre_sudoku(puzzle)):
        afficher_sudoku(puzzle,puzzle_original)
    else:
        print("Pas de solution existe")
```

```python
puzzle = [
    [5, 3, '$', '$', 7, '$', '$', '$', '$'],
    [6, '$', '$', 1, 9, 5, '$', '$', '$'],
    ['$', 9, 8, '$', '$', '$', '$', 6, '$'],
    [8, '$', '$', '$', 6, '$', '$', '$', 3],
    [4, '$', '$', 8, '$', 3, '$', '$', 1],
    [7, '$', '$', '$', 2, '$', '$', '$', 6],
    ['$', 6, '$', '$', '$', '$', 2, 8, '$'],
    ['$', '$', '$', 4, 1, 9, '$', '$', 5],
    ['$', '$', '$', '$', 8, '$', '$', 7, 9]
]
sudoku(puzzle)
```

## Carrés **Latins**
Résoudre le problème du carré latin présente des similitudes avec la résolution d'une grille de Sudoku. Notre algorithme génère les diagonales possibles qui correspondent à la description. Ensuite, il remplit le reste du carré en utilisant la même méthode de remplissage que pour le problème précédent.


### Génération des valeurs diagonales
Nous avons mis en place un générateur pour produire des diagonales potentielles. Le paramètre `limit` permet de spécifier la valeur maximale autorisée. Cela est essentiel pour la récursivité du générateur.

```python
## generator of sequences with size <size> of numbers less then <limit> to somme up to <somme>`
def gen_diag(size,somme,limit="default"):
  if limit=="default":
    limit=size
  if size==1 and somme<=limit and somme>0:
    yield [somme]
  elif size<=1 and somme>limit:
    yield None
  else:
    for i in range(1,min(limit,somme)+1):
      r=[]
      for decomp in gen_diag(size-1,somme-i,i):
        if(decomp is not None):
          r=[i]+decomp
          if(len(r)==size):
            yield r
```

### Remplissage du carré
Les fonctions suivantes sont des versions modifiées des fonctions de la dernière section. Par exemple, `est_valid` n'a plus besoin de vérifier chaque bloc de 3 par 3.



```python
import matplotlib.pyplot as plt

############### MODIFIED TO SOLVE LATIN SQUARE ############

def est_valide(square, ligne, colonne, num): # same as sudoku wihout checking each 3x3 block
    n=len(square)
    ## check ligne
    for y in range(n):
        if square[ligne][y] == num:
            return False
    ## check column
    for x in range(n):
        if square[x][colonne] == num:
            return False

    return True

def fill_square(square): ## same as sudoku (resoudre_sudoku)
    n=len(square)
    for ligne in range(n):
        for colonne in range(n):
            if square[ligne][colonne] == 0:
                for num in range(1, n+1):
                    if est_valide(square, ligne, colonne, num):
                        square[ligne][colonne] = num
                        if fill_square(square):
                            return True
                        square[ligne][colonne] = 0
                return False
    return True

def afficher_latin(puzzle, puzzle_original): ## different to highlight the diagonal(aficher_sudoku)
    n=len(puzzle_original)
    fig, ax = plt.subplots()
    # Set the background color
    ax.set_facecolor('lightgrey')
    # Draw the grid
    for i in range(n+1):
          ax.plot([i, i], [0,n], color='black', linewidth=1)
          ax.plot([0,n], [i,i], color='black', linewidth=1)

    for i in range(n):
        for j in range(n):
            if puzzle[i][j] != 0:
                color, wieght =  ('red','bold') if i==j else ('black','normal')
                ax.text(j+0.5, n-0.5-i, str(puzzle[i][j]), fontweight=wieght,ha='center', va='center', color=color)

    # Add title
    plt.title("Latin square", fontsize=20, pad=20)

    # Remove axis
    ax.axis('off')

    # Show the plot
    plt.show()

def solve_square(square): # return true if a solution is found
    square_original = [row.copy() for row in square]
    if(fill_square(square)):
        afficher_latin(square,square_original)
        return True
    return False

def latin_square(size,somme,stop_at_first_find=True):
  square=[[0 for _ in range(size)] for _ in range(size)]
  found=False

  for sq in gen_diag(size,somme):
    if (len(set(sq))==2):
      continue
    for i in range(len(square)):
      square[i][i]=sq[i]
    if(solve_square(square) and stop_at_first_find):
      print('found a latin square!')
      found = True
      break
    square=[[0 for _ in range(size)] for _ in range(n)]
  if not found :
    print('No latin square fits the desciption')
```

# **2. NumPy: Régression Logistique**


```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_moons, make_circles, make_blobs
```

La classe `LogisticRegression` :

Note : Les methodes `grandient_descent` et `newton_method` sont applés dans la fonction `fit()`

```python

class LogisticRegression:
    #Constructor
    def __init__(self, n_iterations, optimizer='gradient_descent'):
        self.n_iterations = n_iterations
        self.beta = None
        self.alpha = None
        self.optimizer = optimizer
        self.loss = []

    # fit(X,y)
    def fit(self, X, y):
        if self.optimizer == 'gradient_descent':
            self.gradient_descent(X, y)
        elif self.optimizer == 'newton':
            self.newton_method(X, y)

    # method 1  gradient descent
    def gradient_descent(self, X, y):
        n, d = X.shape
        # we add a column of ones to X
        X = np.hstack((np.ones((n, 1)), X))
        # we initialize beta to a vector of zeros
        self.beta = np.zeros(d + 1)
        # we calculate the largest singular value of X
        lambda_ = np.linalg.svd(X, compute_uv=False)[0]
        # we calculate alpha
        self.alpha = 4 / (lambda_ ** 2)
        # we initialize t to 0
        t = 0
        # we iterate until t reaches n_iterations
        while t < self.n_iterations:
            # we calculate the gradient of L
            gradient = X.T.dot(self.sigmoid(X.dot(self.beta)) - y)
            # we update beta
            self.beta = self.beta - self.alpha * gradient
            # convergence speed
            self.loss.append(-np.sum(y * np.log(self.sigmoid(X.dot(self.beta))) + (1 - y) * np.log(1 - self.sigmoid(X.dot(self.beta)))))
            # we increment t
            t = t + 1
        # we return self
        return self

    # method 2 newton method
    def newton_method(self, X, y):
        n, d = X.shape
        # we add a column of ones to X
        X = np.hstack((np.ones((n, 1)), X))
        # we initialize beta to a vector of zeros
        self.beta = np.zeros(d + 1)
        # we calculate the largest singular value of X
        t = 0
        # we iterate until t reaches n_iterations
        while t < self.n_iterations:
            # we calculate the gradient of L
            gradient = X.T.dot(self.sigmoid(X.dot(self.beta)) - y)
            H = X.T.dot(np.diag(self.sigmoid(X.dot(self.beta)) / (1 - self.sigmoid(X.dot(self.beta))))).dot(X)
            self.beta = self.beta - np.linalg.inv(H).dot(gradient)
            # convergence speed
            self.loss.append(-np.sum(y * np.log(self.sigmoid(X.dot(self.beta))) + (1 - y) * np.log(1 - self.sigmoid(X.dot(self.beta)))))
            t = t + 1
        return self

    # predict(X) predicts the labels of the data X.
    def predict(self, X):
        X = np.hstack((np.ones((X.shape[0], 1)), X))
        # we return the predicted labels
        return np.round(self.sigmoid(X.dot(self.beta)))

    # sigmoid(x) calculates the sigmoid of x.
    def sigmoid(self, x):
        # we return the sigmoid of x
        return 1 / (1 + np.exp(-x))
```

# 3. Arbres de **Décision :**
construction d'un arbre de decision

**le constructeur:** prend comme argument gini ou etropy pour la methode de calcule d'impurte construit un arbre vide.

**_calculate_impurity:** methode privee pour calculee l'impurtee soit gini ou entropy

**_split_data :** Cette méthode privée divise les données en deux sous-ensembles en fonction du meilleur split propose par **_find_best_split**.

**_find_best_split :** Cette méthode privée trouve le meilleur split pour les données en minimisant l’impureté. Elle parcourt toutes les caractéristiques et toutes les valeurs uniques de chaque caractéristique pour trouver le split qui minimise l’impureté.

**_build_tree :** Cette méthode privée construit l’arbre de décision en utilisant une approche récursive. Elle continue à diviser les données en utilisant **_split_data** jusqu’à ce que toutes les données d’un nœud appartiennent à la même classe ou qu’il n’y ait plus de caractéristiques à diviser.

**fit :** Cette méthode entraîne l’arbre de décision sur un ensemble de données d’entraînement. Elle construit l’arbre de décision en utilisant les méthodes privées **_build_tree** et **_find_best_split**.

**_predict_single :** Cette méthode privée prédit la classe d’une seule instance en utilisant l’arbre de décision.

**predict :** Cette méthode prédit la classe de plusieurs instances en utilisant la méthode privée **_predict_single**.

**_plot_tree_recursive et plot_tree :** Ces méthodes sont utilisées pour visualiser l’arbre de décision. La méthode _plot_tree_recursive est une méthode privée qui visualise l’arbre de décision en utilisant une approche récursive. La méthode plot_tree est la méthode publique que vous pouvez appeler pour visualiser l’arbre.

**decision_boundary:**cette methode detecte est dessine la frontiere de decision

```python
import pandas as pd
import matplotlib.pyplot as plt


class DecisionTree:
    def __init__(self, criterion="gini"):
        self.criterion = criterion
        self.tree = None

    def _calculate_impurity(self, y):
        classes = {label: len(group) for label, group in pd.DataFrame(y).groupby(0)}
        total_samples = len(y)
        if self.criterion == "gini":
            impurity = 1 - sum((classes[v] / total_samples) ** 2 for v in classes)
        elif self.criterion == "entropy":
            impurity = -sum((classes[v] / total_samples) * (sum(classes[v] / total_samples) / sum(1 / total_samples)) for v in classes)
        else:
            raise ValueError("Invalid criterion. Choose 'gini' or 'entropy'")
        return impurity

    def _split_data(self, X, y, feature, value):
        mask = X[feature] <= value
        return X[mask], y[mask], X[~mask], y[~mask]

    def _find_best_split(self, X, y):
        best_feature, best_value, best_impurity = None, None, float('inf')
        for feature in X.columns:
            values = X[feature].unique()
            for value in values:
                X_left, y_left, X_right, y_right = self._split_data(X, y, feature, value)
                impurity = (len(y_left) * self._calculate_impurity(y_left) +
                            len(y_right) * self._calculate_impurity(y_right)) / len(y)
                if impurity < best_impurity:
                    best_feature, best_value, best_impurity = feature, value, impurity
        return best_feature, best_value

    def _build_tree(self, X, y):
        if len(set(y)) == 1:
            return y[0]
        if X.empty:
            return max(set(y), key=list(y).count)

        best_feature, best_value = self._find_best_split(X, y)
        if best_feature is None:
            return max(set(y), key=list(y).count)

        tree = {'feature': best_feature, 'value': best_value, 'left': None, 'right': None}
        X_left, y_left, X_right, y_right = self._split_data(X, y, best_feature, best_value)
        tree['left'] = self._build_tree(X_left, y_left)
        tree['right'] = self._build_tree(X_right, y_right)

        return tree

    def fit(self, X, y):
        if not isinstance(X, pd.DataFrame):
            X = pd.DataFrame(X)

        if isinstance(y, pd.Series):
            y = y.values

        self.tree = self._build_tree(X, y)

    def _predict_single(self, tree, sample):
        if isinstance(tree, (int, np.int64, float, str)):  # Handle leaf nodes
            return tree
        feature, value = tree['feature'], tree['value']
        if sample[feature] <= value:
            return self._predict_single(tree['left'], sample)
        else:
            return self._predict_single(tree['right'], sample)

    def predict(self, X):
          return [self._predict_single(self.tree, sample) for sample in X]


    def _plot_tree_recursive(self, ax, node, depth, x_offset, y_offset, spacing):
        if isinstance(node, (int, np.int64, float, str)):  # Handle leaf nodes
            ax.text(x_offset, y_offset, str(node), fontsize=12, ha='center', va='center',
                    bbox=dict(boxstyle='round', facecolor='white'))
        else:
            ax.text(x_offset, y_offset, f"{node['feature']} <= {node['value']:.2f}", fontsize=12, ha='center',
                    va='center', bbox=dict(boxstyle='round', facecolor='white'))
            ax.plot([x_offset, x_offset - spacing], [y_offset - 1, y_offset - 2], linewidth=1, color='black')
            ax.plot([x_offset, x_offset + spacing], [y_offset - 1, y_offset - 2], linewidth=1, color='black')

            self._plot_tree_recursive(ax, node['left'], depth + 1, x_offset - spacing, y_offset - 2, spacing / 2)
            self._plot_tree_recursive(ax, node['right'], depth + 1, x_offset + spacing, y_offset - 2, spacing / 2)

    def plot_tree(self):
        if self.tree is None:
            print("Decision Tree not trained yet.")
            return None

        fig, ax = plt.subplots(figsize=(30, 12))
        self._plot_tree_recursive(ax, self.tree, 1, 0, 0, 1)
        ax.set_title("Custom Decision Tree Visualization")
        return fig

    def decision_boundary(self, X, y=None):
        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

        xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
        Z = np.array(self.predict(np.c_[xx.ravel(), yy.ravel()])).reshape(xx.shape)

        cmap_background = ListedColormap(['#FFAAAA', '#AAAAFF'])  # Light colors for decision regions
        cmap_points = ListedColormap(['#FF0000', '#0000FF'])  # Dark colors for data points

        plt.contourf(xx, yy, Z, cmap=cmap_background, alpha=0.3)

        if y is not None:
            plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_points, edgecolor='k', s=40)
        else:
            plt.scatter(X[:, 0], X[:, 1], cmap=cmap_points, edgecolor='k', s=40)

        plt.title("Decision Boundary of Custom Decision Tree")
        plt.xlabel("Feature 1")
        plt.ylabel("Feature 2")
        plt.show()


```

```python
from sklearn.datasets import make_moons, make_circles, make_blobs
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

# Create datasets
datasets = [
    make_moons(noise=0.3, random_state=0),
    make_circles(noise=0.2, factor=0.5, random_state=1),
    make_blobs(n_samples=100, centers=2, n_features=2, center_box=(0, 20), random_state=0)
]

# Visualize DecisionTree and RandomForest for each dataset
for idx, (X, y) in enumerate(datasets):
    print(f"\nDataset {idx + 1}:")

    # Create and train a DecisionTree
    dt = DecisionTree(criterion="gini")
    dt.fit(X, y)

    # Visualize DecisionTree
    plt.figure(figsize=(10, 6))
    dt.plot_tree()
    plt.title(f'Decision Tree Visualization - Dataset {idx + 1}')
    plt.show()

```

# **Forêt Aléatoire**
tout dabord la classe **RandomForest** herite de la classe **decisionTree**.


__init__ : Cette méthode initialise la forêt d’arbres de décision. Elle prend trois arguments : n_estimators qui spécifie le nombre d’arbres dans la forêt, criterion qui spécifie le critère à utiliser pour mesurer la qualité d’un split, et max_features qui spécifie le nombre de caractéristiques à considérer lors de la recherche du meilleur split. Le critère peut être “gini” ou “entropy”. Par défaut, n_estimators est 100, criterion est “gini”, et max_features est None (ce qui signifie que toutes les caractéristiques sont considérées).

**_build_tree_recursive :** Cette méthode privée construit un arbre de décision en utilisant une approche récursive. Elle continue à diviser les données jusqu’à ce que toutes les données d’un nœud appartiennent à la même classe ou que la profondeur maximale soit atteinte. Si max_features est spécifié, un sous-ensemble aléatoire des caractéristiques est sélectionné avant de chercher le meilleur split.

**_build_tree :** Cette méthode privée construit un arbre de décision en utilisant la méthode _build_tree_recursive. Elle convertit d’abord X en DataFrame . Ensuite, elle crée un échantillon bootstrap des données d’entraînement et utilise cet échantillon pour construire l’arbre.

**fit :** Cette méthode entraîne la forêt d’arbres de décision sur un ensemble de données d’entraînement. Elle construit chaque arbre sur un échantillon bootstrap des données d’entraînement.

**_predict_single_tree :** Cette méthode privée prédit la classe d’une seule instance en utilisant un arbre de décision en utilisent les methode **predict_singl**e .

**predict :**Cette méthode prédit la classe de plusieurs instances en utilisant la méthode privée **_predict_single_tree**.

**plot_forest :** cette methode utilise les methode _plot_tree et _plot_tree-recursive pour visualiser tout les arbre construit dans la foret

```python
import pandas as pd
import matplotlib.pyplot as plt

class RandomForest(DecisionTree):
    def __init__(self, n_estimators=100, criterion="gini", max_features=None):
        super().__init__(criterion=criterion)
        self.n_estimators = n_estimators
        self.max_features = max_features
        self.trees = []

    def _build_tree_recursive(self, X, y, depth=1):
        if len(set(y)) == 1 or depth == 3:
            return max(set(y), key=list(y).count)
        if X.empty:
            return max(set(y), key=list(y).count)

        if self.max_features:
            selected_features = random.sample(list(X.columns), self.max_features)
            X = X[selected_features]

        best_feature, best_value = self._find_best_split(X, y)
        if best_feature is None:
            return max(set(y), key=list(y).count)

        tree = {'feature': best_feature, 'value': best_value, 'left': None, 'right': None}
        X_left, y_left, X_right, y_right = self._split_data(X, y, best_feature, best_value)
        tree['left'] = self._build_tree_recursive(X_left, y_left, depth + 1)
        tree['right'] = self._build_tree_recursive(X_right, y_right, depth + 1)

        return tree

    def _build_tree(self, X, y):
        if not isinstance(X, pd.DataFrame):
            X = pd.DataFrame(X)

        if isinstance(y, pd.Series):
            y = y.values

        # Random sampling with replacement
        sample_indices = np.random.choice(len(X), len(X), replace=True)
        sample_X, sample_y = X.iloc[sample_indices], y[sample_indices]

        return self._build_tree_recursive(sample_X, sample_y)

    def fit(self, X, y):
        if not isinstance(X, pd.DataFrame):
            X = pd.DataFrame(X)

        if isinstance(y, pd.Series):
            y = y.values

        self.trees = []
        for _ in range(self.n_estimators):
            tree = self._build_tree(X, y)
            self.trees.append(tree)

    def predict(self, X):
        if not isinstance(X, pd.DataFrame):
            X = pd.DataFrame(X)

        return [self._predict_single_tree(self.trees[0], sample) for _, sample in X.iterrows()]

    def _predict_single_tree(self, tree, sample):
        if isinstance(tree, (int, np.int64, float, str)):  # Handle leaf nodes
            return tree
        feature, value = tree['feature'], tree['value']
        if sample[feature] <= value:
            return self._predict_single_tree(tree['left'], sample)
        else:
            return self._predict_single_tree(tree['right'], sample)

    def plot_forest(self):
        if not self.trees:
            print("Random Forest not trained yet.")
            return

        num_trees = len(self.trees)
        num_rows = num_trees // 2 + num_trees % 2  # Adjust the number of rows dynamically
        num_cols = 2

        plt.figure(figsize=(12, 8))
        for i, tree in enumerate(self.trees):
            plt.subplot(num_rows, num_cols, i + 1)
            self._plot_tree_recursive(plt.gca(), tree, 1, 0, 0, 1)
            plt.title(f'Tree {i + 1}')
        plt.suptitle("Random Forest Visualization")
        plt.show()

```

```python
import numpy as np
import cvxpy as cvx
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from sklearn.datasets import make_moons, make_circles, make_blobs
import warnings
warnings.filterwarnings('ignore')

# karnel functions
def linear_kernel(x1, x2, gamma):
    return np.dot(x1.T, x2)

def rbf_kernel(x1, x2,gamma):
        return np.exp(-gamma * np.linalg.norm(x1 - x2) ** 2)

class SVM(object):

    def __init__(self, kernel, C):
        self.kernel = kernel
        self.C = C
        self.gamma = None

        if self.C is not None:
            self.C = float(self.C)

    # generating the kernel matrix
    def _kernel_transform(self, X_train, kernel):
        if kernel == 'linear':
            kfunc = linear_kernel
        elif kernel == 'rbf':
            self.gamma = 1 / (2 * np.var(X_train))
            kfunc = rbf_kernel

        self.kfunc = kfunc

        n_samples, n_features = X_train.shape
        self.n_samples = n_samples
        self.n_features = n_features

        K = np.zeros((n_samples, n_samples))

        for i in range(n_samples):
            for j in range(n_samples):
                K[i,j] = kfunc(X_train[i], X_train[j], self.gamma)

        return K

    # quadratic problem
    def quad_solver(self, K_matrix, y_train, C):
        n = y_train.shape[0]
        y_= np.diag(y_train)
        alpha = cvx.Variable(n)
        # constraints
        constraints = []
        for i in range(n):
            constraints += [alpha[i] >= 0, alpha[i] <= C]
        constraints += [y_train * alpha == 0]
        self.model_obj = cvx.Maximize(np.ones(n) * alpha - 0.5* cvx.quad_form(alpha, np.outer(y_train, y_train) * K_matrix, assume_PSD=True))#convex
        #quad form
        self.model = cvx.Problem(objective=self.model_obj, constraints=constraints)
        self.model.solve()


        #support vector
        self.a = np.array(alpha.value).flatten()
        self.sv = self.a > 1e-5
        self.ind = np.arange(len(self.a))[self.sv]
        self.a = self.a[self.sv]
        self.sv_y = y_train[self.sv]
        self.sv_x= self.X_train[self.sv]

        # intercept
        self.b = 0
        for n in range(len(self.a)):
            self.b+=self.sv_y[n]
            self.b-=np.sum(self.a*self.sv_y*K_matrix[self.ind[n],self.sv])
        self.b/=len(self.a) #average

        #weight vector
        if self.kernel == 'linear':
            self.w = np.zeros(self.n_features)
            for n in range(len(self.a)):
                    self.w += self.a[n] * self.sv_y[n] * self.sv_x[n]
        elif self.kernel == 'rbf':
            self.w = None

    # fit function
    def fit(self, X_train, y_train):
        self.X_train = X_train
        K = self._kernel_transform(X_train, self.kernel)
        self.K = K
        self.quad_solver(K, y_train, self.C)

    # predict function
    def predict(self, X):
        if self.w is not None:

            return np.sign(np.dot(X,self.w) + self.b)

        else:
            y_predict = np.zeros(len(X))#

            for i in range(len(X)):
                s=0
                for a, sv_y, sv_x in zip(self.a, self.sv_y, self.sv_x):
                    s += a * sv_y * self.kfunc(X[i], sv_x, self.gamma)
                y_predict[i] = s

            return np.sign(y_predict + self.b)

# ploting function
def plot_decision_boundary(X, y, model, ax, title):
    x1 = np.linspace(np.min(X[:,0])-0.5, np.max(X[:,0])+0.5, 100)
    x2 = np.linspace(np.min(X[:,1])-0.5, np.max(X[:,1])+0.5, 100)
    x1, x2 = np.meshgrid(x1, x2)
    yp = np.zeros(x1.shape)
    for i in range(x1.shape[0]):
        yp[i] = model.predict(np.c_[x1[i], x2[i]])
    cmap = mcolors.LinearSegmentedColormap.from_list("n",['#ff0000','#0000ff'])
    ax.contourf(x1, x2, yp, alpha=0.5, cmap=cmap)
    scatter = ax.scatter(X[:,0], X[:,1], c=y, s = 50, edgecolor='k', cmap=cmap)
    ax.set_title(title, fontsize=14)
```

```python
datasets = [
    make_moons(noise=0.3, random_state=0),
    make_circles(noise=0.2, factor=0.5, random_state=1),
    make_blobs(n_samples=100, centers=2, n_features=2, center_box=(0, 20), random_state=0)
]

methods = ['linear', 'rbf']

fig, axs = plt.subplots(len(datasets), len(methods), figsize=(12, 12))

for i, dataset in enumerate(datasets):
    X, y = dataset
    y = np.where(y==0, -1, y)
    for j, method in enumerate(methods):
        if method == 'linear':
            svm = SVM(kernel=method, C=0.025)
        else:
            svm = SVM(kernel=method, C=np.inf)
        svm.fit(X, y)
        y_pred = svm.predict(X)
        plot_decision_boundary(X, y, svm, axs[i, j], f'Data set {i+1}, {method} kernel')

plt.tight_layout()
plt.show()
```
